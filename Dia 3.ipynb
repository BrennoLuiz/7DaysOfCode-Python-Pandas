```python
import pandas as pd
import seaborn as sns
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt
```

```python
dados_2010_1 = pd.read_csv('dados_emprestimos/emprestimos-20101.csv')
dados_2010_2 = pd.read_csv('dados_emprestimos/emprestimos-20102.csv')
dados_2011_1 = pd.read_csv('dados_emprestimos/emprestimos-20111.csv')
dados_2011_2 = pd.read_csv('dados_emprestimos/emprestimos-20112.csv')
dados_2012_1 = pd.read_csv('dados_emprestimos/emprestimos-20121.csv')
dados_2012_2 = pd.read_csv('dados_emprestimos/emprestimos-20122.csv')
dados_2013_1 = pd.read_csv('dados_emprestimos/emprestimos-20131.csv')
dados_2013_2 = pd.read_csv('dados_emprestimos/emprestimos-20132.csv')
dados_2014_1 = pd.read_csv('dados_emprestimos/emprestimos-20141.csv')
dados_2014_2 = pd.read_csv('dados_emprestimos/emprestimos-20142.csv')
dados_2015_1 = pd.read_csv('dados_emprestimos/emprestimos-20151.csv')
dados_2015_2 = pd.read_csv('dados_emprestimos/emprestimos-20152.csv')
dados_2016_1 = pd.read_csv('dados_emprestimos/emprestimos-20161.csv')
dados_2016_2 = pd.read_csv('dados_emprestimos/emprestimos-20162.csv')
dados_2017_1 = pd.read_csv('dados_emprestimos/emprestimos-20171.csv')
dados_2017_2 = pd.read_csv('dados_emprestimos/emprestimos-20172.csv')
dados_2018_1 = pd.read_csv('dados_emprestimos/emprestimos-20181.csv')
dados_2018_2 = pd.read_csv('dados_emprestimos/emprestimos-20182.csv')
dados_2019_1 = pd.read_csv('dados_emprestimos/emprestimos-20191.csv')
dados_2019_2 = pd.read_csv('dados_emprestimos/emprestimos-20192.csv')
dados_2020_1 = pd.read_csv('dados_emprestimos/emprestimos-20201.csv')
```

```python
emprestimos_df = pd.concat([dados_2010_1, dados_2010_2, dados_2011_1, dados_2011_2, dados_2012_1, dados_2012_2, dados_2013_1,
                            dados_2013_2, dados_2014_1, dados_2014_2, dados_2015_1, dados_2015_2, dados_2016_1, dados_2016_2,
                            dados_2017_1, dados_2017_2, dados_2018_1, dados_2018_2, dados_2019_1, dados_2019_2, dados_2020_1],
                            ignore_index=True)

emprestimos_df
```

```python
emprestimos_df.head()
```

```python
exemplares_df = pd.read_parquet('dados_exemplares.parquet')
exemplares_df
```

```python
emprestimos_final = emprestimos_df.merge(exemplares_df)
emprestimos_final
```

```python
emprestimos_final = emprestimos_final.drop_duplicates()
emprestimos_final.head()
```

```python
emprestimos_final = emprestimos_final.drop(columns=['registro_sistema'])
emprestimos_final.head()
```

```python
emprestimos_final['matricula_ou_siape'] = emprestimos_final['matricula_ou_siape'].astype(str)
emprestimos_final.head()
```

```python
lista_CDU = []
for CDU in emprestimos_final['localizacao']:
    if(CDU < 100):
        lista_CDU.append('Generalidades, Ciência e conhecimento')
    elif(CDU < 200):
        lista_CDU.append('Filosofia e psicologia')
    elif(CDU < 300):
        lista_CDU.append('Religião')
    elif(CDU < 400):
        lista_CDU.append('Ciências sociais')
    elif(CDU < 500):
        lista_CDU.append('Classe vaga. Provisoriamente não ocupada')
    elif(CDU < 600):
        lista_CDU.append('Matemática e ciências naturais')
    elif(CDU < 700):
        lista_CDU.append('Ciências aplicadas')
    elif(CDU < 800):
        lista_CDU.append('Belas artes')
    elif(CDU < 900):
        lista_CDU.append('Linguagem. Língua. Linguística')
    elif(CDU < 1000):
        lista_CDU.append('Geografia. Biografia. História')

emprestimos_final['CDU'] = lista_CDU

emprestimos_final.head()
```

```python
emprestimos_final['id_emprestimo'].value_counts()
```

```python
emprestimos = len(emprestimos_final['id_emprestimo'].drop_duplicates())
emprestimos
```

```python
exemplares = len(emprestimos_final)
exemplares
```

```python
emprestimos_ano = pd.DataFrame(emprestimos_final['data_emprestimo'].value_counts()).reset_index()
emprestimos_ano.columns = ['anos', 'quantidade']
emprestimos_ano['anos'] = pd.to_datetime(emprestimos_ano['anos'])
emprestimos_ano
```

```python
emprestimos_ano.loc[:,'anos'] = emprestimos_ano.loc[:,'anos'].dt.year
```

```python
emprestimos_ano = emprestimos_ano.value_counts().groupby(['anos']).sum().to_frame('quantidade').sort_values(by='anos', ascending=True).reset_index()
emprestimos_ano
```

```python
# plt.figure()
# emprestimos_ano.plot(x="anos", y="quantidade")

sns.set_theme(context='notebook',
              style='darkgrid',
              palette='deep',
              font_scale=1.3,
              rc={"figure.figsize":(15,8)})
```

```python
emp_ano = sns.lineplot(data=emprestimos_ano, x='anos', y='quantidade')
emp_ano.set(xlabel=None, ylabel=None)
emp_ano.tick_params(axis='x', rotation=30)
emp_ano.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))

emp_ano.set_title('Quantidade de exemplares emprestados por ano' + '\n', size=20, loc='left', weight='bold')

emp_ano = emp_ano
```

```python
emprestimos_mes = pd.DataFrame(emprestimos_final['data_emprestimo'].value_counts()).reset_index()
emprestimos_mes.columns = ['mes', 'quantidade']
emprestimos_mes['mes'] = pd.to_datetime(emprestimos_mes['mes'])
emprestimos_mes
```

```python
emprestimos_mes.loc[:,'mes'] = emprestimos_mes.loc[:,'mes'].dt.month
```

```python
emprestimos_mes = emprestimos_mes.value_counts().groupby(['mes']).sum().to_frame('quantidade').sort_values(by='mes', ascending=True).reset_index()
emprestimos_mes
```

```python
conversao_meses = {0:' ',1:'Jan',2:'Fev',3:'Mar',4:'Abr',5:'Mai',6:'Jun',7:'Jul',8:'Ago',9:'Set',10:'Out',11:'Nov',12:'Dez'}
```

```python
emprestimos_mes.index = emprestimos_mes.index.map(conversao_meses)
```

```python
emprestimos_mes
```

```python
# plt.figure()
# emprestimos_mes.plot(x="mes", y="quantidade")

emp_mes = sns.lineplot(data=emprestimos_mes, x='mes', y='quantidade')
emp_mes.set(xlabel=None,ylabel=None)
emp_mes.set_xticks(range(len(conversao_meses)))
emp_mes.set_xticklabels(['','Jan','Fev','Mar','Abr','Mai','Jun','Jul','Ago','Set','Out','Nov','Dez'])
emp_mes.tick_params(axis='x', rotation=30)
emp_mes.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))
emp_mes.set_title('Quantidade de exemplares emprestados por mês'+'\n', size=20, loc='left', weight='bold')
emp_mes.text(s='Período de 2010 a 2020',x=0.025,y=265000,fontsize=18, ha='left',color='gray')

emp_mes=emp_mes
```

```python
emprestimos_hora = pd.DataFrame(emprestimos_final['data_emprestimo'].value_counts()).reset_index()
emprestimos_hora.columns = ['horas', 'quantidade']
emprestimos_hora['horas'] = pd.to_datetime(emprestimos_hora['horas'])
emprestimos_hora
```

```python
emprestimos_hora.loc[:,'horas'] = emprestimos_hora.loc[:,'horas'].dt.hour
```

```python
emprestimos_hora = emprestimos_hora.value_counts().groupby(['horas']).sum().to_frame('quantidade').sort_values(by='horas', ascending=True).reset_index()
emprestimos_hora
```

```python
emp_hora = sns.barplot(data=emprestimos_hora, x='horas', y='quantidade',
                       palette='Blues', hue='quantidade', dodge=False)
plt.legend([],[], frameon=False)

emp_hora.set(xlabel='Horário', ylabel=None)
emp_hora.tick_params(axis='x', rotation=30)
emp_hora.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))
emp_hora.set_title('Quantidade de exemplares emprestados por faixa horária' + '\n', size=20, loc='left', weight='bold')
emp_hora.text(s='Período de 2010 a 2020',x=-0.4,y=225000,fontsize=18, ha='left',color='gray')
emp_hora = emp_hora
```
